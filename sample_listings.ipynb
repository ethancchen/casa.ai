```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
DATA_PATH = Path().resolve() / "data" / "listings.csv"
assert DATA_PATH.exists()
```

```python
np.random.seed(42)
```

```python
df = pd.read_csv(DATA_PATH)
```

## Shrink dataset

```python
df_sample = df.sample(5)
df_sample
```

```python
df_sample["room_type"]
```

```python
cols_to_keep_long_str = "listing_url, name, summary, space, description, neighborhood_overview, notes, access, house_rules, property_type, room_type, host_about, host_location, neighbourhood_cleansed, accommodates, bathrooms, bedrooms, beds, bed_type, amenities, square_feet, price, weekly_price, monthly_price, security_deposit, cleaning_fee, guests_included"
```

```python
cols_to_keep = cols_to_keep_long_str.split(", ")
```

```python
df_sample = df_sample[cols_to_keep]
```

```python
df_sample.to_csv(DATA_PATH.with_name("sampled_listings.csv"), index=False)
```
